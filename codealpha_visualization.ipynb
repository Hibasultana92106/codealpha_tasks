# Import Required Libraries

The following libraries are used for data analysis and visualization:
- Pandas
- Matplotlib
- Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the Dataset

The Netflix dataset is loaded using Pandas for further analysis and visualization.

# Dataset Overview

This step helps us understand:
- Number of rows and columns
- Data types
- Missing values
- Structure of the dataset

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

# 🎬 Movies vs TV Shows

This visualization compares the number of Movies and TV Shows available on Netflix.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

# Top Countries Producing Content

This graph shows which countries contribute the most content to Netflix.

```python
plt.figure(figsize=(10,5))

df['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Shows")

plt.show()
```

# Netflix Content Growth

This visualization shows how Netflix content has increased over the years.

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(10,5))

df['year_added'].value_counts().sort_index().plot(kind='line')

plt.title("Netflix Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

# Most Popular Genres

This graph highlights the most common genres available on Netflix.

```python
plt.figure(figsize=(12,6))

df['listed_in'].value_counts().head(10).plot(kind='bar')

plt.title("Top Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")

plt.show()
```

# Missing Values Analysis

This heatmap helps identify missing or null values in the dataset.

```python
plt.figure(figsize=(10,5))

sns.heatmap(df.isnull(), cbar=False)

plt.title("Missing Values Heatmap")

plt.show()
```

# Data Story

The analysis reveals that Netflix experienced rapid growth after 2015.
Movies are more common than TV Shows on the platform.
The USA and India contribute most of the content.
Drama and International genres are highly popular among viewers.
