# Netflix Dataset — Exploratory Data Analysis

## 1. Objective
The aim of this analysis is to explore the Netflix dataset to understand content distribution, identify patterns over time, and evaluate data quality.

## 2. Research Questions

- What is the distribution of Movies and TV Shows on Netflix?  
- Which countries contribute the most content?  
- How has content production changed over the years?  
- What are the most common content ratings?  
- Are there any missing or duplicate values in the dataset? 

## 3. Data Loading
The dataset is loaded using pandas.

```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
```

# 4. Understanding the Data
The dataset structure is explored using basic functions.

```python
df.head()
df.shape
df.columns
df.info()
```

Observation:  
The dataset contains multiple columns such as type, title, country, release year, and rating. It includes both categorical and numerical data.

```python
df.isnull().sum()
```

## 5. Data Quality Check
The dataset is checked for missing values and duplicates.

```python
df.fillna("Unknown", inplace=True)
```

Observation:
Some columns contain missing values. Duplicate records are checked to ensure consistency.

```python
df.duplicated().sum()
```

## 6. Data Cleaning
Missing values are handled to improve data quality.

```python
df['type'].value_counts()
df['country'].value_counts().head(10)
df['release_year'].value_counts().sort_index()
df['rating'].value_counts()
```

Observation:
Most content falls under TV-MA and TV-14 categories.

## 7. Hypothesis Testing

**Hypothesis:**  
Netflix produces more Movies than TV Shows.

**Method:**  
Compare frequency of content types.

**Result:**  
The hypothesis is supported as Movies are more frequent than TV Shows.

## 8. Conclusion

The analysis shows that Netflix has more Movies than TV Shows, with the United States being the major contributor. Content production has increased after 2015, and most content is targeted toward mature audiences. Missing data was handled during preprocessing to ensure consistency.
